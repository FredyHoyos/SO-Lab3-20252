```python
#! Importar librerías necesarias
import matplotlib.pyplot as plt   # Librería para generar gráficos
import pandas as pd               # Librería para manejar datos en tablas (DataFrame)

# Datos experimentales obtenidos al ejecutar el programa con diferentes números de hilos
# N_hilos: cantidad de hilos utilizados
# Tp_segundos: tiempo de ejecución paralelo medido para cada cantidad de hilos
data = {
    "N_hilos": [1, 2, 4, 8],
    "Tp_segundos": [0.330342, 0.168281, 0.084310, 0.043990]
}

# Crear un DataFrame (tabla) con los datos anteriores para facilitar los cálculos y análisis
df = pd.DataFrame(data)

# Calcular el Speedup y la Eficiencia para cada número de hilos

# Ts: tiempo de ejecución serial (cuando N = 1 hilo)
Ts = df.loc[0, "Tp_segundos"]

# Speedup = Ts / Tp → mide cuánto más rápido se ejecuta el programa en paralelo
df["Speedup"] = Ts / df["Tp_segundos"]

# Eficiencia = Speedup / N → mide qué tan bien se aprovechan los recursos (procesadores)
df["Eficiencia"] = df["Speedup"] / df["N_hilos"]

# Mostrar el DataFrame resultante con las columnas:
# N_hilos, Tp_segundos, Speedup y Eficiencia
df

```

```python
# Crear una nueva figura para el gráfico, con un tamaño de 8x5 pulgadas
plt.figure(figsize=(8, 5))

# Dibujar la gráfica de Speedup en función del número de hilos
# marker='o' → usa círculos para marcar los puntos
# linestyle='-' → conecta los puntos con una línea continua
# linewidth=2 → grosor de la línea
plt.plot(df["N_hilos"], df["Speedup"], marker='o', linestyle='-', linewidth=2)

# Título del gráfico
plt.title("Gráfico de Speedup vs Número de Hilos")

# Etiqueta del eje X
plt.xlabel("Número de Hilos (N)")

# Etiqueta del eje Y
plt.ylabel("Speedup (Ts / Tp)")

# Activar la cuadrícula para facilitar la lectura de los valores
plt.grid(True)

# Mostrar el gráfico en pantalla
plt.show()

```

```python
# Crear una nueva figura para el gráfico, con un tamaño de 8x5 pulgadas
plt.figure(figsize=(8, 5))

# Dibujar la gráfica de Eficiencia en función del número de hilos
# marker='s' → usa cuadrados como marcadores
# color='orange' → define el color de la línea y los puntos
# linewidth=2 → establece el grosor de la línea
plt.plot(df["N_hilos"], df["Eficiencia"], marker='s', color='orange', linewidth=2)

# Título del gráfico
plt.title("Eficiencia vs Número de Hilos")

# Etiqueta del eje X
plt.xlabel("Número de Hilos (N)")

# Etiqueta del eje Y
plt.ylabel("Eficiencia (Speedup / N)")

# Activa la cuadrícula para una mejor visualización de los datos
plt.grid(True)

# Muestra el gráfico en pantalla
plt.show()
```
